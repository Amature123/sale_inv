### All the asosorment are follow the OR operation in de Morgan law

```python
Classification.head()
```

```python
months={'AP':4,'AU':8,'DE':12}
def split_year_month(name):
    p=name[:2]
    year=2000+int(p)
    s=name[2:]
    for i in months.keys():
        if s==i:
            s=months.get(i)
    season=s+year*100
    return season
Classification['sales_season']=Classification['sales_season'].apply(lambda x : split_year_month(x))
Classification['lauch_season']=Classification['lauch_season'].apply(lambda x : split_year_month(x) if x!='KHÁC' else 0)
```

```python
Classification.tail()
```

### Notice 4 season column , first 2 number of each Value are stood for Year after 2000 , and the last 2 letter are stood for 3 months of year (APril , AUgust , DEcember) , so we could try to split the data into datetime type . The reason is it could be some relatives with other datasets

```python
date_sale=Classification['sales_season'].apply(lambda x: pd.to_datetime(str(x),format='%Y%m'))
Classification['sales_season']=date_sale.dt.year
Classification['sales_season_num']=date_sale.dt.month
Classification.head()
```

```python
Classification.rename(columns={"sales_season": "year_sale", "sales_season_num": "month_sale"})
```

```python
early=datetime(1,1,1)
date_lauch=Classification['lauch_season'].apply(lambda x: pd.to_datetime(str(x),format='%Y%m') if x!=0 else early)
```

```python
date_lauch = pd.to_datetime(date_lauch, errors='coerce')
Classification['lauch_season']=date_lauch.dt.year
Classification['lauch_season_num']=date_lauch.dt.month
Classification.head()
```
