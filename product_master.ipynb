### check discount between listing price and cost price

```python
df['discount']=df['listing_price']/df['cost_price']
df['discount'].describe()
```

something off : cause listing price is always higher than cost price . but in this column appear that listing price lower than cost price . so we have tune again the cost price to exact price . By devide listing price with the median of discount column

```python
#Remove outlier
upper_lim = df['discount'].quantile(.95)
lower_lim = df['discount'].quantile(.05)

df = df[(df['discount'] < upper_lim) & (df['discount'] > lower_lim)]
df['discount'].describe()
```

```python
df['cost_price']=df['listing_price'] / df['discount'].mean()
```

### Feature Adding

```python
df.launch_season.value_counts()
```

```python
from datetime import datetime
```

```python
months={'AP':4,'AU':8,'DE':12,'BM':5} #BM: Business month
def split_year_month(name):
    p=name[:2]
    year=2000+int(p)
    s=name[2:]
    for i in months.keys():
        if s==i:
            s=months.get(i)
    season=s+year*100
    return season
df['launch_season']=df['launch_season'].apply(lambda x : split_year_month(x) if x!='KHÁC' else 0)
```

```python
early=datetime(1,1,1)
date_lauch=df['launch_season'].apply(lambda x: pd.to_datetime(str(x),format='%Y%m') if x!=0 else early)
```

```python
date_lauch = pd.to_datetime(date_lauch, errors='coerce')
df['year_launch']=date_lauch.dt.year
df['month_launch']=date_lauch.dt.month
df.drop('discount',axis=1,inplace=True)
```
